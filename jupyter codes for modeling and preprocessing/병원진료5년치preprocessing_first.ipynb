```python
import pandas as pd
import numpy as np
```

```python
yr = 2016
```

```python
df = pd.read_csv('da_'+str(yr)+'.csv', encoding='euc-kr')
```

```python
df
```

```python
eye = df[df['주상병코드'].str.startswith('H041')]
```

```python
eye = eye[['시도코드', '요양개시일자', '성별코드', '연령대코드(5세단위)', '주상병코드', '부상병코드']]
```

```python
res = eye.pivot_table(values=[], index=['시도코드', '요양개시일자'], columns=['성별코드'],
                     aggfunc=len, fill_value=0)
```

```python
res2 = eye.pivot_table(values=[], index=['시도코드', '요양개시일자'], columns=['연령대코드(5세단위)'],
               aggfunc=len, fill_value=0)
```

```python
fin = pd.concat([res, res2], axis=1)
```

```python
names = ['성별코드=1', '성별코드=2']
names.extend(['연령대코드='+str(x) for x in range(1, 19)])
fin.columns = names
```

```python
fin['전체빈도'] = fin['성별코드=1'] + fin['성별코드=2']
```

```python
fin = fin.reset_index()
```

```python
fin.to_csv('병원진료'+str(yr)+'.csv', encoding='euc-kr', index=False)
```
