```python
import pandas as pd
```

```python
def go(yr, D):
    df1 = pd.read_excel('/Users/leejunki/weather/'+str(yr)+'/'+'_1.xlsx', encoding='euc-kr')
    df2 = pd.read_excel('/Users/leejunki/weather/'+str(yr)+'/'+'_2.xlsx', encoding='euc-kr')
    df3 = pd.read_excel('/Users/leejunki/weather/'+str(yr)+'/'+'_3.xlsx', encoding='euc-kr')
    df4 = pd.read_excel('/Users/leejunki/weather/'+str(yr)+'/'+'_4.xlsx', encoding='euc-kr')
    
    res1 = df1.groupby([df1['지역'], df1['측정소코드'], df1['측정소명'], df1['측정일시']//100],
                       as_index=False).mean()
    res2 = df2.groupby([df2['지역'], df2['측정소코드'], df2['측정소명'], df2['측정일시']//100],
                       as_index=False).mean()
    res3 = df3.groupby([df3['지역'], df3['측정소코드'], df3['측정소명'], df3['측정일시']//100],
                       as_index=False).mean()
    res4 = df4.groupby([df4['지역'], df4['측정소코드'], df4['측정소명'], df4['측정일시']//100],
                       as_index=False).mean()
    
    res1['일시'] = res1['측정일시'] // 100
    res2['일시'] = res2['측정일시'] // 100
    res3['일시'] = res3['측정일시'] // 100
    res4['일시'] = res4['측정일시'] // 100
    
    cols = res1.columns.tolist()
    cols = cols[:3] + [cols[-1]] + cols[3:-1]
    res1 = res1[cols]

    cols = res2.columns.tolist()
    cols = cols[:3] + [cols[-1]] + cols[3:-1]
    res2 = res2[cols]

    cols = res3.columns.tolist()
    cols = cols[:3] + [cols[-1]] + cols[3:-1]
    res3 = res3[cols]

    cols = res4.columns.tolist()
    cols = cols[:3] + [cols[-1]] + cols[3:-1]
    res4 = res4[cols]
    
    res1['일시'] = res1['일시'].astype('int')
    del res1['측정일시']

    res2['일시'] = res2['일시'].astype('int')
    del res2['측정일시']

    res3['일시'] = res3['일시'].astype('int')
    del res3['측정일시']

    res4['일시'] = res4['일시'].astype('int')
    del res4['측정일시']
    
    res = pd.concat([res1, res2, res3, res4])
    res.groupby([res['지역'], res['일시']], as_index=False).mean()
    data = res.groupby([res['지역'], res['일시']], as_index=False).mean()
    data['지역명'] = data.apply(lambda row: row['지역'][:2], axis=1)
    data.groupby([data['지역명'], data['일시']], as_index=False).mean()
    temp = data.groupby([data['지역명'], data['일시']], as_index=False).mean()
    
    del temp['측정소코드']
    temp = temp.replace({"지역명": D})
    print(temp)
    temp.to_csv('/Users/leejunki/weather/finedust'+str(yr)+'.csv', encoding='euc-kr')
```
