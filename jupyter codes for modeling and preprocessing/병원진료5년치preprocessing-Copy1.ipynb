```python
import pandas as pd
import numpy as np
```

```python
yr = 2016
```

```python
df = pd.read_csv('da_'+str(yr)+'.csv', encoding='euc-kr')
```

```python
df
```

```python
eye = df[df['주상병코드'].str.startswith('H041')]
```

```python
eye = eye[['시도코드', '요양개시일자', '주상병코드', '가입자일련번호']]
```

```python
res = eye.pivot_table(values=[], index=['시도코드', '요양개시일자'], columns=['주상병코드'], aggfunc=len, fill_value=0)
```

```python
res.to_csv('병원진료'+str(yr)+'.csv', encoding='euc-kr', index=False)
```
